```python
from sagemaker import get_execution_role
import sagemaker as sage

sess = sage.Session()
role = get_execution_role()
```

```python
# Lets set up our SageMaker notebook instance for local mode.
!/bin/bash ./utils/setup.sh
```

```python
%%sh

# The name of our algorithm
algorithm_name=sagemaker-recsys-graph

# cd container

# chmod +x kggraph/train
# chmod +x kggraph/serve

account=$(aws sts get-caller-identity --query Account --output text)

# Get the region defined in the current configuration (default to us-west-2 if none defined)
region=$(aws configure get region)
region=${region:-us-west-1}

fullname="${account}.dkr.ecr.${region}.amazonaws.com/${algorithm_name}:latest"

# If the repository doesn't exist in ECR, create it.

aws ecr describe-repositories --repository-names "${algorithm_name}" > /dev/null 2>&1

if [ $? -ne 0 ]
then
    aws ecr create-repository --repository-name "${algorithm_name}" > /dev/null
fi

# Get the login command from ECR and execute it directly
$(aws ecr get-login --region ${region} --no-include-email)
```

```python
from sagemaker.estimator import Estimator

hyperparameters = {'train-steps': 100}

instance_type = 'local_gpu'
# instance_type = 'ml.p3.2xlarge'

estimator = Estimator(role=role,
                      instance_count=1,
                      instance_type=instance_type,
                      image_name='sagemaker-recsys-graph:latest',
                      output_path="s3://{}/output".format(sess.default_bucket()),
                      image_uri='002224604296.dkr.ecr.us-east-1.amazonaws.com/sagemaker-recsys-graph',
                      hyperparameters=hyperparameters)

# estimator = Estimator(role=role,
#                       instance_count=1,
#                       instance_type=instance_type,
#                       image_name='sagemaker-decision-trees:latest',
#                       output_path="s3://{}/output".format(sess.default_bucket()),
#                       image_uri='002224604296.dkr.ecr.us-east-1.amazonaws.com/sagemaker-decision-trees',
#                       hyperparameters=hyperparameters)

# estimator.fit()
```

```python
# train
estimator.fit()
```

```python
# serve
from sagemaker.predictor import json_serializer
result = estimator.deploy(1, instance_type, serializer=json_serializer)
```
